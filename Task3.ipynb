```python
file.drop(columns = "Province/State" , axis=1, inplace=True)
file.head()
```

```python
filtered_file_zeros = file[file['Active'] > 0]
filtered_file_zeros.head()
```

```python
filtered_file_negative = file[file['Active'] < 0]
filtered_file_negative.head()
```

```python
file.loc[file['Active'] == 0, 'Active'] = pd.NA
file.loc[file['Active'] < 0, 'Active'] = pd.NA
file.iloc[:10]
```

```python
file_with_nan = file 
file_with_nan.isnull().head(10)
```

```python
file_with_drop = file_with_nan.dropna()
file_with_drop.iloc[:10]
```

```python
file_with_drop.info()
```

```python
file_with_drop["Active"] = file_with_drop["Active"].astype(int)
file_with_drop["Active"].dtypes
```

```python
print("\nCategorical Columns Summary:")
print(file_with_drop.describe(include=['object']))
```

```python
sorted_by_active = file_with_drop.sort_values(by='Active', ascending=False)
sorted_by_active.iloc[:10]
```

```python
file_with_drop.to_csv('covid_19.csv', index=False)
print("Filtered data saved to co.csv.")
```
